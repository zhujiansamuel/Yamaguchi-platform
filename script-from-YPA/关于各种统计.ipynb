```python
from datetime import datetime, timedelta, timezone

JST = timezone(timedelta(hours=9))
start = datetime(2025,10,30, 7, 15, 0, tzinfo=JST)
end   = datetime(2025,11,9, 13, 11, 0, tzinfo=JST)

minutes = int((end - start).total_seconds() // 60)  # 59040
timestamps = [(start + timedelta(minutes=i)).isoformat(timespec="seconds")
              for i in range(minutes - 1, -1, -1)]

len(timestamps), timestamps[:3], timestamps[-3:]

```

```python
import json
import subprocess, time
url = "https://verbless-sadistically-jayceon.ngrok-free.dev/AppleStockChecker/purchasing-time-analyses/dispatch_ts/"


for i, ts in enumerate(timestamps, 1):
    payload = json.dumps({"timestamp_iso": ts,})
    out = subprocess.check_output(
        ["curl", "-sS", "-X", "POST", url,
         "-H", "Content-Type: application/json",
         "-d", payload],
        stderr=subprocess.STDOUT
    ).decode("utf-8", "replace")
    print(f"[{i}/{len(timestamps)}] {ts}\n{out}\n")

    if i%100 == 0:
        time.sleep(60)
    time.sleep(0.3)
```
