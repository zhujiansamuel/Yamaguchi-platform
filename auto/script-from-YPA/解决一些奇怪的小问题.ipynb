```python
from wcwidth import wcswidth
def east_asian_width_pad(s, total_width, align='left', fillchar=' '):
    """
    根据东亚字符实际显示宽度进行对齐填充
    align: 'left', 'right', 'center'
    """
    current_width = wcswidth(s)  # 计算实际显示宽度
    if current_width == -1:  # 包含不可打印字符等特殊情况
        current_width = len(s)

    padding = total_width - current_width
    if padding <= 0:
        return s  # 已超过宽度，直接返回

    if align == 'left':
        return s + fillchar * padding
    elif align == 'right':
        return fillchar * padding + s
    elif align == 'center':
        left_pad = padding // 2
        right_pad = padding - left_pad
        return fillchar * left_pad + s + fillchar * right_pad
items = [
    "Hello",
    "こんにちは",
    "世界",
    "Pythonプログラミング",
    "ABC"
]

for item in items:
    print(east_asian_width_pad(item, 20, align='left') + "| データ")
```
